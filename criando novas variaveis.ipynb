```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv' , sep=';')
```

```python
dados
```

```python
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
```

```python
dados
```

```python
dados['Valor m2'] = (dados['Valor'] / dados['Area']).round(2)
```

```python
dados
```

```python
dados['Valor Bruto m2'] = (dados['Valor Bruto'] / dados['Area']).round(1)
```

```python
dados
```

```python
casa = ['Casa', 'Casa ou Condominio', 'Casa de Vila']
```

```python
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: "Casa" if x in casa else 'Apartamento')
```

```python
dados
```

```python
dados_aux = pd.DataFrame(dados[['Tipo Agregado','Valor Bruto', 'Valor m2', 'Valor Bruto m2']])
```

```python
del dados_aux['Valor Bruto']
```

```python
dados_aux
```

```python
dados.drop(['Valor Bruto', 'Valor Bruto m2'], axis = 1, inplace = True )
```

```python
dados_aux.pop('Valor Bruto m2')
```

```python
dados_aux
```

```python
dados.drop(dados_aux, axis = 1, inplace = True)
```

```python
dados
```

```python
dados.to_csv('dados/aluguel_residencial.csv', sep=';', index = False)
```
